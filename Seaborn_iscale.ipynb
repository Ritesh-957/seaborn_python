```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
var1 = [1,2,3,4,5,6,7,8]
var2 = [4,3,5,3,6,7,9,2]
```

```python
plt.plot(var1, var2, color = 'r')
plt.title("Matplot Lib Plot")
plt.show()
```

# Lineplot

```python
sns.lineplot(x = var1, y = var2)
plt.title(" Seaborn Library Line Plot")
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame({"var1":var1, "var2":var2})
df
```

```python
sns.lineplot(x = 'var1' , y = 'var2', data = df)
plt.show()
```

```python
df2 = pd.read_csv("penguins.csv")
df1 = df2.head(20)
df1
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = df1)
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = df1, hue = "sex")
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = df1, hue = "sex",style = "sex", palette = 'rocket_r', markers =["o",">"])
plt.grid()
plt.title("Line Graph")
plt.show()
```

# Order Parameter

```python
order1= ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = 'island' , y = 'bill_length_mm' , data =  df2, hue = 'sex', order = order1)
plt.show()
```

# Hue Order

```python
order2 = ["FEMALE","MALE"]
sns.barplot(x = 'island' , y = 'bill_length_mm' , data =  df2, hue = 'sex', hue_order = order2, order = order1)
plt.show()
```

# Colors

```python
sns.barplot(x = 'island', y = 'bill_length_mm' , data = df2, hue = 'sex', hue_order = order2, palette = "prism", order = order1)
plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm' , data = df2, hue = 'sex', hue_order = order2, palette = "gist_heat_r", order = order1)
plt.show()
```

# Saturation

```python
sns.barplot(x = 'island', y = 'bill_length_mm' , data = df2, hue = 'sex', hue_order = order2, palette = "gist_heat_r", order = order1, saturation = 0.2)
plt.show()
```
